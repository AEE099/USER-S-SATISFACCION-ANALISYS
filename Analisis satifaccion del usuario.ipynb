```python
#Converting specific colums to numeric types
#This is necessary for numerical analysis and visualization

df_numerico= ["Cantidad Edad", "Cantidad Sexo", "Cantidad Nivel Académico",
              "Cantidad Tiempo de Espera del procedimiento","Cantidad Tiempo de Entrega de Resultados",
              "Cantidad Evaluación del servicio", "Cantidad Información Proporcionada",
              "Cantidad infraestructura", "Cantidad Satisfacción General"]
df[df_numerico]= df[df_numerico].apply (pd.to_numeric, errors="coerce")

#Verifying the data types after conversion 

print("Data types after conversion: \n", df.dtypes)
```

```python
# Contar la frecuencia de cada comentario adicional (ignorando mayúsculas/minúsculas y espacios)
comentarios = df["Comentarios Adicionales que desee agregar:"].str.strip().str.lower()
tabla_frecuencia = comentarios.value_counts().reset_index()
tabla_frecuencia.columns = ["Comentario", "Frecuencia"]
tabla_frecuencia.index += 1  # Para enumerar desde 1

print(tabla_frecuencia)

plt.figure(figsize=(10, 6))
sns.barplot(
    y="Comentario",
    x="Frecuencia",
    data=tabla_frecuencia,
    palette="mako"
)
plt.title("Frecuencia de Comentarios Adicionales")
plt.xlabel("Frecuencia")
plt.ylabel("Comentario")
plt.tight_layout()
plt.show()
```
